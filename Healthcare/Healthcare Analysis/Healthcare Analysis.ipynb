# Healthcare Data Analysis

```python
SELECT *
FROM 'healthcare_dataset.csv'
ORDER BY "Date of Admission" DESC;
```

```python
-- 1. What are the top 5 insurance providers?
SELECT "insurance provider", COUNT("insurance provider") AS count_providers
FROM 'healthcare_dataset.csv'
GROUP BY "insurance provider"
ORDER BY count_providers DESC;
```

```python
-- 2. What is the average billing amount per customer?
SELECT name, ROUND(AVG("Billing Amount"),2) AS avg_billing_customer
FROM 'healthcare_dataset.csv'
GROUP BY name
ORDER BY name;
```

```python
-- 3. What is the most common medical condition? What are the 10 most common medical conditions?
SELECT "Medical Condition", COUNT("Medical Condition") AS count_of_conditions
FROM 'healthcare_dataset.csv'
GROUP BY "Medical Condition"
ORDER BY count_of_conditions DESC;
```

```python
-- 4. List the top 5 patients with the highest billing amount. No duplicates.
SELECT DISTINCT name, ROUND("Billing Amount", 2)
FROM 'healthcare_dataset.csv'
ORDER BY "Billing Amount" DESC
LIMIT 5;
```

```python
-- 5. On average, how many admissions are there for each admission type?
SELECT DISTINCT "Admission type"
FROM 'healthcare_dataset.csv';
```

```python
-- 6. Return the name, age, gender, and medical condition of patients who went to Hunter-Grant hospital.
SELECT name, age, gender, "medical condition"
FROM 'healthcare_dataset.csv'
WHERE Hospital = 'Hunter-Grant';
```

```python
-- 7. What is the average billing amount of each medical condition?
SELECT "Medical Condition", ROUND(AVG("Billing Amount"),2) AS avg_billing_amount
FROM 'healthcare_dataset.csv'
GROUP BY "Medical Condition";
```

```python
-- 9. Which insurance provider has the highest billing amount on average? What are the most common medical conditions for this provider (top 10)?
WITH top_provider AS (
	SELECT "Insurance Provider", ROUND(AVG("Billing Amount"),2) AS avg_billings
	FROM 'healthcare_dataset.csv'
	GROUP BY "Insurance Provider"
	ORDER BY avg_billings DESC
	LIMIT 1)


SELECT "Medical Condition", COUNT ("Medical Condition") AS most_common_condition
FROM 'healthcare_dataset.csv'
WHERE "Insurance Provider" = (SELECT "Insurance Provider" FROM top_provider)
GROUP BY "Medical Condition"
ORDER BY most_common_condition DESC
LIMIT 10;
```

```python
-- 11. Which blood type most commonly has diabetes?
SELECT "Blood Type", COUNT(*) AS diabetes_cases
FROM 'healthcare_dataset.csv'
WHERE "Medical Condition" = 'Diabetes'
GROUP BY "Blood Type"
ORDER BY diabetes_cases DESC
LIMIT 1;
```

```python
-- 12. Which hospital most commonly deals with the insurance provider Cigna?
SELECT "Hospital", COUNT (*) AS provider_cigna
FROM 'healthcare_dataset.csv'
WHERE "Insurance Provider" = 'Cigna'
GROUP BY "Hospital"
ORDER BY provider_cigna DESC
LIMIT 1;
```

```python
-- 13. Provide one query to answer a question about this data set that is not already listed.
-- Which Medication had the highest sales volume in 2024?
SELECT DISTINCT "Medication", ROUND(SUM("Billing Amount"),1) AS highest_sales_volume
FROM 'healthcare_dataset.csv'
WHERE "Date of Admission" BETWEEN '2024-01-01T00:00:00.000' AND '2024-05-07T00:00:00.000'
GROUP BY "Medication"
ORDER BY highest_sales_volume DESC;

```
